```python
def log_bin_data(qs, values, log_n, errors=None):
    qs = np.array(qs)
    values = np.array(values)
    log_bins = np.logspace(np.log10(qs.min()), np.log10(qs.max()), log_n + 1)
    bin_centers = (log_bins[:-1] + log_bins[1:]) / 2
    
    binned_values = np.zeros(log_n)
    binned_errors = np.zeros(log_n) if errors is not None else None

    for i in range(log_n):
        bin_mask = (qs >= log_bins[i]) & (qs < log_bins[i + 1])
        bin_data = values[bin_mask]
        if len(bin_data) > 0:
            binned_values[i] = np.mean(bin_data)
            if errors is not None:
                bin_errors = np.array(errors)[bin_mask]
                binned_errors[i] = np.sqrt(np.sum(bin_errors**2)) / len(bin_data)
    
    return (bin_centers, binned_values, binned_errors) if errors is not None else (bin_centers, binned_values)

def add_colorbar(ax, vec, label=None, location='top'):
    vec = np.array(vec)
    divider = make_axes_locatable(ax)

    size = '2%'  # Reduced size for the colorbar
    pad = 0.3    # Padding between the axis and the colorbar

    if location == 'top':
        cax = divider.append_axes('top', size=size, pad=pad)
        orientation = 'horizontal'
        cax.set_position([0.2, 0.9, 0.6, 0.03])  # Adjust position [left, bottom, width, height]
    elif location == 'bottom':
        cax = divider.append_axes('bottom', size=size, pad=pad)
        orientation = 'horizontal'
        cax.set_position([0.2, 0.1, 0.6, 0.03])  # Adjust position [left, bottom, width, height]
    elif location == 'left':
        cax = divider.append_axes('left', size=size, pad=pad)
        orientation = 'vertical'
        cax.set_position([0.1, 0.2, 0.03, 0.6])  # Adjust position [left, bottom, width, height]
    elif location == 'right':
        cax = divider.append_axes('right', size=size, pad=pad)
        orientation = 'vertical'
        cax.set_position([0.9, 0.2, 0.03, 0.6])  # Adjust position [left, bottom, width, height]
    else:
        raise ValueError("location must be one of 'top', 'bottom', 'left', 'right'")

    cmap = mpl.cm.jet  # Colormap from blue to red
    norm = mpl.colors.Normalize(vmin=vec.min(), vmax=vec.max())

    cb1 = mpl.colorbar.ColorbarBase(cax, cmap=cmap, norm=norm, orientation=orientation)
    cb1.set_label(label, fontsize=12, labelpad=5)

    # Adjust label orientation
    if location in ['left', 'right']:
        cb1.set_label(label, fontsize=12, labelpad=5, rotation=270)
    else:
        cb1.set_label(label, fontsize=12, labelpad=5, rotation=0)
```

```python
import matplotlib as mpl

fig2, (ax2,ax3) = plt.subplots(1,2, figsize =(15,7))

# Initialize variables
cumulative_time = 0
exp_times = []
Qs = []
Q_errors = []
# absorber = [9,9,9,9,9,9,9,9,9,9,9,9,9,9,9]
absorber = 9

q_range = (0.01, 0.3)
count = 0

qs = s2.Xana.get_item(1)['soq'][:,0] # get q values from first entry in 'ind', assuming all entries were analysed with the same q ring setup
mask = (qs >= q_range[0]) & (qs <= q_range[1])
qs = qs[mask]

# Make sure the length of temperature data matches the number of SAXS entries
temperature_data = np.array([24,63,110,110,110,110,110,110,110,110,24,24,24,24,24])

# Process each SAXS entry
ax2.set_prop_cycle(plt.cycler('color',plt.cm.jet(np.linspace(0, 1, int(len(ind_saxs))   ))))#8))))

for i in ind_saxs:
    current_exp_time = d2.db['t_exposure'][i] * (d2.db['last'][i])
    current_delay_time = d2.db['t_delay'][i] * (d2.db['last'][i])
    cumulative_time += current_exp_time + current_delay_time
    exp_time = cumulative_time
    exp_times.append(exp_time)
    
    ISaxs = (s2.Xana.get_item(i)['soq'][:,1]/d2.db['t_exposure'][i])/0.732046 ** (absorber)   
    VSaxs = (s2.Xana.get_item(i)['soq'][:,2]/ d2.db['t_exposure'][i])/0.732046 ** (absorber)
    ISaxs_sub = ISaxs - ISaxs_water #*(0.73615**(16-8)) # average intensity normalised attenuation to abs16  E = 8.54 keV
    VSaxs_sub = VSaxs - VSaxs_water #*(0.73615**(16-8))#/(0.732046**24) # variance in saxs (error)
    
    # # Filter q and Iq to be within the specified q_range
    
    ISaxs_sub = ISaxs_sub[mask]
    VSaxs_sub = VSaxs_sub[mask]

    # Log binning
    qs_n, ISaxs_n, VSaxs_n = log_bin_data(qs, ISaxs_sub, log_n, errors=VSaxs_sub)
    
    # Calculate scattering invariant Q
    Q = simps(qs_n**2 * ISaxs_n, qs_n)
    Q_err = simps(qs_n**2 * VSaxs_n, qs_n)

    Qs.append(Q)
    Q_errors.append(Q_err)
    
    # Plot I(q) vs q for the current SAXS entry
    ax2.errorbar(qs_n, ISaxs_n, VSaxs_n, label=f'{exp_time:.2f} s')
    
    count+=1

# Convert lists to numpy arrays
exp_times = np.array(exp_times)
Qs = np.array(Qs)
Q_errors = np.array(Q_errors)


# Plot Q vs waiting time
ax3.errorbar(exp_times, Qs, yerr=Q_errors, fmt='o-')
ax3.set_xscale('log')
ax3.set_yscale('log')
ax3.set_xlabel('Waiting Time (s)')
ax3.set_ylabel('Scattering Invariant Q')
ax3.set_title('Q vs Waiting Time')

# Set plot labels and legends
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel('$q$ $[nm^{-1}]$')
ax2.set_ylabel('$I(q)$ $[photon/s]$')
# ax2.legend(loc='best', fontsize=10)
# ax2.set_title(f'{sample[:6]} SAXS Profiles')
# ax2.set_xlim(qs[0],qs[-1])
# Add the colorbar
# add_colorbar(ax2, exp_times, 'Waiting Time (s)')
add_colorbar(ax2,exp_times,r'$t_\mathrm{w}$ (s)',location='top') #exposure times(s)

# Add second y-axis for temperature
# vx = np.linspace(qs[0],qs[-1], len(temperature_data))
ax2_twin = ax2.twiny()
# ax2_twin = ax2.twinx()
ax2_twin.plot(exp_times, temperature_data, 'k-', label='Temperature (°C)')  # Plot an invisible line to set up the scale
# ax2_twin.set_ylabel('Temperature (°C)', color='k')
# ax2_twin.set_ylim(0, 120)
# ax2_twin.set_xlim(24, 100)
# ax2_twin.tick_params(axis='x', colors='k')
ax2_twin.set_xticklabels([]) 

# plt.savefig(f'{sample}-Q{q_range} updated.png')
plt.tight_layout()
plt.show()
```
